```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Extra√ß√£o dos Dados

Os dados foram disponibilizados em formato JSON, simulando uma API armazenada em um reposit√≥rio GitHub.
Cada registro representa um cliente, contendo informa√ß√µes demogr√°ficas, servi√ßos contratados e dados financeiros.


```python
with open('TelecomX_Data.json') as f:
    data = json.load(f)

df = pd.json_normalize(data)
df.head()
```

## Transforma√ß√£o dos Dados

Realiza-se a limpeza, padroniza√ß√£o e convers√£o de tipos para viabilizar a an√°lise.


```python
df.columns = df.columns.str.replace('.', '_')
```

## An√°lise Explorat√≥ria (EDA)

```python
sns.countplot(x='Churn', data=df)
plt.show()
```

**Insight:** Clientes com contrato mensal apresentam maior taxa de churn.


## Conclus√£o

Os resultados indicam que o churn est√° associado principalmente a contratos mensais,
menor tempo de perman√™ncia e cobran√ßas mensais elevadas.
